```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df
```

```python
sns.countplot(df, y='survived')
plt.title('countof survived vs not survived')
plt.show()
```

```python
sns.countplot(df, x='survived', hue='sex')
plt.title('countof survived vs not survived')
plt.show()
```

```python
sns.barplot(df, x = 'class', y='survived', palette='coolwarm')
plt.title('survival Rate by class')
plt.show()
```

```python
sns.histplot(df, x='age', bins=30, kde=True, color='red')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
# 2. Boxplot of age distribution by class
sns.boxplot(data=df, x='class', y='age', palette='Set3')
plt.title('Age Distribution by Class')
plt.show()
```

```python
# 3. Violin plot of age distribution by survival status
sns.violinplot(data=df, x='survived', y='age', hue='sex', split=True, palette='muted')
plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
# 4. Pairplot of numerical features
sns.pairplot(df, hue='survived', palette='husl')
plt.title('Pairplot of Numerical Features')
plt.show()
```

```python
# 6. Swarmplot of fare distribution by class
sns.swarmplot(data=df, x='class', y='fare', hue='sex', palette='deep')
plt.title('Fare Distribution by Class')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and an axes
fig, ax = plt.subplots()

# Plot the data with different styles
ax.plot(x, y, color='red', linestyle='--', marker='o', label='Line 1')
ax.plot(x, y + 1, color='blue', linestyle=':', marker='x', label='Line 2')
ax.plot(x, y + 2, color='green', linestyle='-.', marker='^', label='Line 3')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Line Graph with Different Styles')

# Add a legend
ax.legend()

# Add a grid
ax.grid(True)

# Show the plot
plt.show()
```

```python
import plotly.express as px
```

```python
px.scatter(df, x='age', y='fare', color='survived', title='age vs fare with survival',
           template='plotly_dark')
```

```python
px.sunburst(df,path=['pclass','sex','survived'],
            values='age',title='survival by class and gender',width=500,height=500,
template='plotly_dark', color='age', color_continuous_scale='RdBu',)
```

```python
px.histogram(df, x='age', nbins=30,color='survived',
             title='age distribution with survival',
             template='plotly_dark')
```

```python
px.treemap(df, path=['pclass','survived'], values='age',
           title='survival by class', width=800, height=500,
           template='plotly_dark', color='age', color_continuous_scale='RdBu',)
```

```python
survived_count = df['survived'].value_counts()
px.pie(names=survived_count.index,values=survived_count.values,
       title='survival Rate', template='plotly_dark',
       width=500, height=500,)
```

```python
import matplotlib.pyplot as plt

# Bar chart of survival count
survived_count = df['survived'].value_counts()
plt.bar(survived_count.index, survived_count.values, color=['red', 'green'])
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.title('Survival Count')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot of age vs fare with survival
fig = px.scatter(df, x='age', y='fare', color='survived', 
                 title='Age vs Fare with Survival', 
                 template='plotly_dark')
fig.show()
```

```python
# 2. Bar Chart: Survival Rate by Class
fig = px.bar(df, x='pclass', y='survived', color='pclass', 
              title='Survival Rate by Class', 
              template='plotly_dark', barmode='group')
fig.show()
```

```python
fig = px.histogram(df, x='age', nbins=30, color='sex',
                   title='age distribution by Gender',
                   template='plotly_dark')
fig.show()
```

```python
# 5. Violin Plot: Age Distribution by Survival
fig = px.violin(df, x='survived', y='age', color='sex', 
                 box=True, points='all', 
                 title='Age Distribution by Survival', 
                 template='plotly_dark')
fig.show()
```

```python
# 7. Sunburst Chart: Survival by Class and Gender
fig = px.sunburst(df, path=['pclass', 'sex', 'survived'], 
                   values='fare', 
                   title='Survival by Class and Gender', 
                   template='plotly_dark', 
                   color='age', color_continuous_scale='RdBu')
fig.show()
```

```python
# 9. Line Chart: Fare vs Age
fig = px.line(df.sort_values(by='age'), x='age', y='fare', 
               title='Fare vs Age', 
               template='plotly_dark')
fig.show()
```

```python
# 10. Density Heatmap: Age vs Fare
fig10 = px.density_heatmap(df, x='age', y='fare', z='survived', 
                           title='Density Heatmap of Age vs Fare', 
                           template='plotly_dark', 
                           color_continuous_scale='Viridis')
fig10.show()
```
