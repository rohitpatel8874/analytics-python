```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df
```

```python
sns.countplot(df, y='survived')
plt.title('countof survived vs not survived')
plt.show()
```

```python
sns.countplot(df, x='survived', hue='sex')
plt.title('countof survived vs not survived')
plt.show()
```

```python
sns.barplot(df, x = 'class', y='survived', palette='coolwarm')
plt.title('survival Rate by class')
plt.show()
```

```python
sns.histplot(df, x='age', bins=30, kde=True, color='red')
plt.title('Age Distribution of Passengers')
plt.show()
```
